```python
#DISTANCE TO FOR EACH GEOHASH TO A SPECIFIC COORDINATE
def distance_from_geohashes(x_coord, y_coord):
    coords_2 = (x_coord, y_coord)
    arr = []
    for each in np.arange(len(stp)):
        coords_1 = (stp.iloc[each][1], stp.iloc[each][2])
        arr.append(geopy.distance.vincenty(coords_1, coords_2).km)
    return(arr)

print("MIN:" , min(distance_from_geohashes(40, -83)))
print("MAX:" , max(distance_from_geohashes(40, -83)))
```

```python
#I WANT TO PUT EACH PARKING TICKET INTO A HASH

dist0 = []
hash_arr = []

for each in np.arange(len(tix)):
    lat_tix = tix.iloc[each][2]
    long_tix = tix.iloc[each][3]
    
    minimum = 100000
    hash_ = ""
    for each2 in np.arange(len(stp)):
    
        lat_stp = stp.iloc[each2][1]
        long_stp = stp.iloc[each2][2]

        coords_1 = (lat_tix, long_tix)
        coords_2 = (lat_stp, long_stp)

        distancia = geopy.distance.vincenty(coords_1, coords_2).km
        
        minimum = min(distancia, minimum)
        if minimum == distancia:
            hash_ = long_stp = stp.iloc[each2][0]
    hash_arr.append(hash_)
    dist0.append(distancia)
```

```python
tix["nearHash"] = hash_arr
tix["HashDist"] = dist0
tix
```
