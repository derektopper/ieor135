```python
#CLOSEST Regions
import math

def five_closest_regions(lat,long, table):
    x1 = lat
    y1 = long
    
    dist_arr = []
    met_arr = []
    
    for each in np.arange(len(table)):
        x2 = table["avgX"][each]
        y2 = table["avgY"][each]
        dist_arr.append(math.hypot(x2 - x1, y2 - y1))
        met_arr.append(table["meter"][each])
      
    inds = []
    distances = []
    for dists in np.sort(dist_arr)[0:5]:
        inds.append(met_arr[dist_arr.index(dists)])
        distances.append(dists)
    return inds, distances
```

```python
def get_results(long,lat, month, day_of_week, time):
    
    
    def get_closest_probs(long,lat, month, day_of_week, time):
        closest_regions = five_closest_regions(long,lat,avg_loc)
        closest_regions = [('prob '+i) for i in closest_regions]
        columns=closest_regions+['Months', 'DayOfWeek', 'TimeOnDay']
        return master[columns][master['Months']==month][master['DayOfWeek']==day_of_week][master['TimeOnDay']==time]
    
    
    res = get_closest_probs(long,lat, month, day_of_week, time)
    for region, prob in zip(res[res.columns.difference(['Months','DayOfWeek','TimeOnDay'])].columns,res[res.columns.difference(['Months','DayOfWeek','TimeOnDay'])].values[0]):
        proba = [i for i in res[res.columns.difference(['Months','DayOfWeek','TimeOnDay'])].values[0] if i != 0]
        if len(proba)==0:
            this = region
        if prob == min(proba):
            this = region
    reg = this.strip('prob ')
    result_long = avg_loc[avg_loc['meter']==reg]['avgX']
    result_lat = avg_loc[avg_loc['meter']==reg]['avgY']
    return [result_long,result_lat,reg]
```
