```python
def centi_sec_to_time(centis):
    secs = centis//100
    hours = secs//(60*60*24)
    minutes = (secs - hours*60*60*24)//60
    seconds = (secs - hours*60*60*24 - minutes*60)
    return str(int(hours))+':'+str(int(minutes))+':'+str(int(seconds))
```

```python
def milli_sec_to_time(millis):
    secs = millis//1000
    hours = secs//(60*60*24)
    minutes = (secs - hours*60*60*24)//60
    seconds = (secs - hours*60*60*24 - minutes*60)
    return str(int(hours))+':'+str(int(minutes))+':'+str(int(seconds))
```

```python
import json
```

```python
with open('parkingMeters.json') as f:
    data = json.load(f)
```

```python
data
```

```python
json_df_format = {}
for meter in data['features']:
    attributes = meter.pop('attributes')
    geometry = meter.pop('geometry')
    attributes['LATITUDE'] = geometry['y']
    attributes['LONGITUDE'] = geometry['x']
    meter_id = attributes.pop('METER_ID')
    json_df_format[meter_id] = attributes

with open('meterToDF.json', 'w') as fp:
    json.dump(json_df_format, fp)
meters = pd.read_json('meterToDF.json', orient='index')
meters.head()
```
